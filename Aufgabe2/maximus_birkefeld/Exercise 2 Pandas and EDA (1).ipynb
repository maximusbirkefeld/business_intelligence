# The Pandas DataFrame

One of the main python libraries for data science and machine learning is `pandas`. The `pandas.DataFrame` is a useful tool for representing, exploring and analysing data in a python program. It is a special type of list that can hold data in columns and rows but also saves the name of every column it contains.

`Pandas` also provides functions to read nearly any type of file. Most of the common formats of data can be read directly through specific functions. For example `pd.read_sql()` or `pd.read_excel()`.
When you get a new DataFrame to work with, the first thing you need to do is explore it and see what it contains. There are several useful methods and attributes for this. If you want to know more about all the different data formats `Pandas` can read, you can look them up in the <a href="https://pandas.pydata.org/pandas-docs/stable/reference/io.html">Input/Output Documentation</a>. In general, the detailed <a href="https://pandas.pydata.org/docs/">Pandas Documentation</a> can be very helpful when running into problems with its components.

- `.head()` returns the first few rows (the “head” of the DataFrame).
- `.info()` shows information on each of the columns, such as the data type and number of missing values.
- `.shape` returns the number of rows and columns of the DataFrame.
- `.describe()` calculates a few summary statistics for each column.

- `.values` : A two-dimensional NumPy array of values.

# Subsetting

When working with data, you may not need all of the variables in your dataset. Square brackets (`[]`) can be used to select only the columns that matter to you in an order that makes sense to you. To select only `"col_a"` of the DataFrame `df`, use `df["col_a"]`. To select `"col_a"` and `"col_b"` of `df`, use `df[["col_a", "col_b"]]`. The second pair of square bracktes indicates a list, as you already know from the last appointment. It means that you can also use a list `cols = ["col_a","col_b",...]` to subset the DataFrame `df` by referencing it in squared brackets `df[cols]`.

A large part of data science is about finding interesting bits in the analysed dataset. Matching some criteria to a subset of rows is good way to start dissecting datasets. This is sometimes known as *filtering rows* or *selecting rows*. There are many ways to subset a DataFrame, perhaps the most common is to use relational operators to return `True` or `False` for each row, then pass that inside square brackets. With a DataFrame `dogs` that could look something like this:

`dogs[dogs["height_cm"] > 60]` or `dogs[dogs["color"] == "tan"]`

You can also filter for multiple conditions at once by using the "bitwise and" operator, `&` which could look like this on the `dogs` DataFrame: 

`dogs[(dogs["height_cm"] > 60) & (dogs["color"] == "tan")]`

The simplest way of subsetting a DataFrame is to use indexing. Using an index to subset a pandas DataFrame is a little different than using the index of python list. If you want to fetch the value in row three in the very first column of the DataFrame `dogs`. The code would look like this: `dogs.iloc[4][0]`. Note that you need to call the `.iloc` method instead of just using square brackets without any function or method.

# Subsetting Categorical Variables
Subsetting data based on a categorical variable often involves using the "or" operator (`|`) to select rows from multiple categories. 
This can get tedious when you want all states in one of three different regions, for example. 
Instead, use the `.isin()` method, which will allow you to tackle this problem by writing one condition instead of three separate ones.

```
colors = ["brown", "black", "tan"]
condition = dogs["color"].isin(colors)
dogs[condition]
```

# Exploratory Data Analysis

Exploratory Data Analysis (EDA) is an essential part of every Data Science project. In first consequence it will, most of the time lead to some type of transformation of the raw data. But to determine what actions need to be made to get a dataset to the point, where a Data Science model can be applied you need to discover the shortcomings of the raw data. Pandas offers good ways to explore raw data by the functions and methods you already learned and applied in this exercise (e.g. `.describe()` or `.info()`).

Another way of further exploring the dataset is calculating aggregates for the different features of the dataset. This includes measures of location, dispersion and higher statistical moments of the features observed distribution. Pandas implements the mean of a column by calling `.mean()` as well as the standard deviation and the calculation of the features empirical variance by calling `.std()` or `.var()`. Even higher statistical moments like `.skew()` have a dedicated function in pandas DataFrames. If that's not enough you can apply your own, self-written function on a DataFrame with `.agg()`. 

To make certain make certain aspects of the raw data more clear (to yourself or others) it may be helpful to create visualizations. A first step might be to visualize a `barplot` of a feature to make assumptions about its distributions. Another plot that can provide major information, especially about outliers in your data is the `boxplot`. The main package for visualizations in python is `matplotlib`'s `pyplot`. Most of the other famous visualization packages either directly build on `matplotlib`'s funcitonality (e.g. `seaborn` and built-in `pandas.plot`) or are highly inspired by it (e.g. `plotly`).

Let's do some EDA on homelessness.

### Exercise
- Calculate and print the standard deviation and mean of the `individuals` and `state_pop` columns in `homelessness`.
- Calculate a costum aggregate of `family_members` by passing the `interquart_range()` function to the DataFrames `.agg()` function.
- Can you explain what the `interquart_range()` calculates? How can it be applicable in a further exploratory analysis?

````python
#Calculate mean of individuals and state_pop
mean_individuals = homelessness["individuals"].mean()
mean_state_pop = homelessness["state_pop"].mean()

#Print the results
print(mean_individuals)
print(mean_state_pop)

#Empty line for readability
print('')

#Calculate standard deviation of individuals and state_pop
std_individuals = homelessness["individuals"].std()
std_state_pop = homelessness["state_pop"].std()

#Print the results
print(std_individuals)
print(std_state_pop)

#Empty line for readability
print('')

#Custom IQR function
def interquart_range(column):
    return column.quantile(0.75) - column.quantile(0.25)

#Apply the IQR function to family_members using .agg()
inter_quartile_range_fm = homelessness["family_members"].agg(interquart_range)

#Print the result
print(inter_quartile_range_fm)
````

- gibt den Interquartilabstand aus IQR
- wichtig bspw. für die Erstellung eines Boxplots (lower/upper bound)
